# Ridge Regresyon

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline
from sklearn import model_selection
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

# Model

```python
hit = pd.read_csv("Hitters.csv")
df = hit.copy()
df = df.dropna()
dms = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
y = df["Salary"]
X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
ridge_model = Ridge(alpha = 0.1).fit(X_train, y_train)
```

```python
ridge_model.coef_
```

```python
10**np.linspace(10,-2,100)*0.5 
```

```python
lambdalar = 10**np.linspace(10,-2,100)*0.5 

ridge_model = Ridge()
katsayilar = []

for i in lambdalar:
    ridge_model.set_params(alpha = i)
    ridge_model.fit(X_train, y_train) 
    katsayilar.append(ridge_model.coef_) 
    

    
ax = plt.gca()
ax.plot(lambdalar, katsayilar) 
ax.set_xscale('log') 

plt.xlabel('Lambda(Alpha) Deerleri')
plt.ylabel('Katsaylar/Arlklar')
plt.title('Dzenliletirmenin Bir Fonksiyonu Olarak Ridge Katsaylar');
```

# Tahmin

```python
y_pred = ridge_model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

# Model Tuning

```python
lambdalar = 10**np.linspace(10,-2,100)*0.5 
```

```python
lambdalar[0:5]
```

```python
ridge_cv = RidgeCV(alphas = lambdalar, scoring = "neg_mean_squared_error", normalize = True)
```

```python
ridge_cv.fit(X_train, y_train)
```

```python
ridge_cv.alpha_
```

```python
ridge_tuned = Ridge(alpha = ridge_cv.alpha_, normalize = True).fit(X_train,y_train)
```

```python
np.sqrt(mean_squared_error(y_test, ridge_tuned.predict(X_test)))
```
